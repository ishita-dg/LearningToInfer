```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats
import seaborn as sns
import os
import matplotlib.cm as cm
import itertools
import math
```

```python
vals = np.random.normal(0, 1, 100)
```

```python
_, bins, _ = plt.hist(vals, bins = 50, normed = True)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.title("Prior", fontsize = 18)
plt.ylabel("Probability of query objects", fontsize = 18)
plt.xlabel("Query Objects", fontsize = 18)
plt.ylim([0,1])
# plt.show()
plt.savefig("Prior.pdf")
```

```python
vals2 = np.concatenate((vals[vals < 0.6], np.random.uniform(1,2,10)))
```

```python
plt.hist(vals2, bins = bins, normed = True)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.title("Conditioned on uninformative cue object", fontsize = 18)
plt.ylabel("Probability of query objects", fontsize = 18)
plt.xlabel("Query Objects", fontsize = 18)
plt.ylim([0,1])
# plt.show()
plt.savefig("Uninf.pdf")

```
